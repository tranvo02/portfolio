##Question 1
Has user activity or engagement dropped, increased or remained stable? What is the extent of change in user activity/engagement?

```
%sql
--Show event types from Events table 
SELECT DISTINCT events.event_type
FROM events
```

```
%sql
--Show event names from Events table 
SELECT DISTINCT events.event_name
FROM events
```

```
%sql
--Show all daily users
SELECT DATE_TRUNC('day', created_at) AS day, COUNT(*) AS all_users
FROM users 
GROUP BY 1
ORDER BY 1 DESC
```

```
%sql
--Show all weekly users
SELECT DATE_TRUNC('week', created_at) AS week, COUNT(*) AS all_users
FROM users 
GROUP BY 1
ORDER BY 1 DESC
```

```
%sql
--Show weekly users engagement only
SELECT DATE_TRUNC('week', events.occurred_at) AS week, COUNT(DISTINCT events.user_id) AS weekly_users_engagement
FROM events 
WHERE event_type = 'engagement' AND event_name = 'login'
GROUP BY 1
ORDER BY 1
```

```
%sql
--Show the graph of weekly users engagement only
SELECT DATE_TRUNC('week', events.occurred_at) AS week, COUNT(DISTINCT events.user_id) AS weekly_users_engagement
FROM events 
WHERE event_type = 'engagement' AND event_name = 'login'
GROUP BY 1
ORDER BY 1
```

##Question 2
Are there any changes in the three stages of the emails funnel? What changes have you discovered? (Please search online for terms such as sales funnel, marketing funnel to understand the funnel concept).

```
%sql
--Find all available actions from the Action column
SELECT DISTINCT emails.action
FROM emails
```

```
%sql
--Investigate each email activity and compare them to the user engagement 
SELECT DATE_TRUNC('week', events.occurred_at) AS week, COUNT(DISTINCT events.user_id) AS weekly_users_engagement,
SUM (CASE WHEN action IN ('sent_reengagement_email') THEN 1 ELSE 0 END) AS reengagement,
SUM (CASE WHEN action IN ('sent_weekly_digest') THEN 1 ELSE 0 END) AS weekly_digest,
SUM (CASE WHEN action IN ('email_open') THEN 1 ELSE 0 END) AS open,
SUM (CASE WHEN action IN ('email_clickthrough') THEN 1 ELSE 0 END) AS clickthrough
FROM events JOIN emails
ON events.user_id = emails.user_id
WHERE events.event_type = 'engagement' AND events.event_name = 'login' 
GROUP BY 1
ORDER BY 1
```

##Question 3
Are changes in user activity/engagement associated with specific devices or hardware ( mobile phone, tablet computers, desktop computers, etc.)  that customers use to read emails and interact with OMG's web portal?

```
%sql
--Find out all the different devices in the 'device' column 
SELECT DISTINCT events.device
FROM events
```

```
%sql
--Show users engagement for each device category 
SELECT DATE_TRUNC('week', events.occurred_at) AS week, COUNT(DISTINCT events.user_id) AS weekly_users_engagement,
SUM(CASE WHEN device IN('iphone 5s', 'nexus 5', 'amazon fire phone', 'nokia lumia 635', 'samsung galaxy note', 'iphone 5', 'iphone 4s', 'samsung galaxy s4', 'htc one') 
THEN 1 ELSE 0 END) AS phone_category,
SUM(CASE WHEN device IN('kindle fire', 'nexus 10', 'ipad mini', 'lenovo thinkpad', 'windows surface', 'ipad air', 'nexus 7', 'samsumg galaxy tablet') THEN 1 ELSE 0 END) AS tablet_category,
SUM(CASE WHEN device IN('hp pavilion desktop', 'macbook air', 'dell inspiron notebook', 'asus chromebook', 'asus chromebook', 'acer aspire notebook', 'macbook pro', 'mac mini', 'acer aspire desktop', 'dell inspiron desktop') 
THEN 1 ELSE 0 END) AS computer_category
FROM events 
WHERE events.event_type = 'engagement' AND events.event_name = 'login' 
GROUP BY 1
ORDER BY 1
```
